## 3.1 Evolution of Agricultural Labour in Mainland Portugal

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Load the uploaded agricultural_labour_clean.csv file (using the filename from the latest successful upload)
df_labour = pd.read_csv('agricultural_labour_clean.csv')

print('First 5 rows of agricultural_labour_clean.csv:')
display(df_labour.head())
```

### Agricultural Labour Data Table

Here is a table generated from the `agricultural_labour_clean.csv` data.

```python
# Select relevant columns for the table, excluding 'code' and 'geographic_location'
df_labour_table = df_labour[['year', 'Agricultural labour']]

# Rename columns for proper display
df_labour_table = df_labour_table.rename(columns={
    'year': 'Year',
    'Agricultural labour': 'Agricultural Labour'
})

# Create a figure and an axes object
fig_labour_table, ax_labour_table = plt.subplots(figsize=(8, 2.5)) # Adjust figure size as needed

# Hide the axes
ax_labour_table.axis('off')
ax_labour_table.axis('tight')

# Create the table using matplotlib
table_labour = ax_labour_table.table(cellText=df_labour_table.values,
                     colLabels=df_labour_table.columns,
                     loc='center',
                     cellLoc='center')

# Adjust font size and layout for better readability
table_labour.auto_set_font_size(False)
table_labour.set_fontsize(10)
table_labour.scale(1.2, 1.2) # Scale the table

plt.title('Agricultural Labour in Mainland Portugal (1989–2019)', loc='left', fontsize=12, pad=20)

# Save the table as a PNG image
plt.savefig('agricultural_labour_table.png', bbox_inches='tight', dpi=300)
print("The table has been saved as 'agricultural_labour_table.png'")

plt.close(fig_labour_table) # Close the figure to free up memory
```

### Evolution of Agricultural Labour Line Graph

Here is a line graph visualizing the evolution of agricultural labour, with the y-axis starting at 0.

```python
# Set the style for the plot
sns.set_style('whitegrid')

fig_labour_line = plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='Agricultural labour', data=df_labour, marker='o')

plt.title('Evolution of Agricultural Labour in Mainland Portugal (1989–2019)')
plt.xlabel('Year')
plt.ylabel('Agricultural Labour')
plt.xticks(df_labour['year'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(bottom=0) # Ensure y-axis starts at 0
plt.ticklabel_format(style='plain', axis='y') # Display full numbers on the y-axis
plt.tight_layout()

# Save the plot to a file
plt.savefig('agricultural_labour_line_chart.png')
print("The line chart for agricultural labour has been saved as 'agricultural_labour_line_chart.png'")

plt.close(fig_labour_line) # Close the figure to free up memory
```

## 3.2 Changes in Agricultural Area and Land Use

#### 3.2.1 Evolution of Total Agricultural Area

Contrary to the sharp decline observed in agricultural labour, total agricultural area in mainland Portugal remained relatively stable over the 30-year period. Table 3.3 presents the evolution of the total agricultural holdings area between 1989 and 2019.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded agricultural_holdings_area_clean.csv file
df_area = pd.read_csv('agricultural_holdings_area_clean.csv')

# Rename the problematic column to a simpler, unambiguous name
df_area = df_area.rename(columns={'Agricultural holdings area (ha)': 'Agricultural_Holdings_Area_ha'})

print('First 5 rows of agricultural_holdings_area_clean.csv:')
display(df_area.head())
```

```python
# Create a figure and an axes object
fig_area_table, ax_area_table = plt.subplots(figsize=(8, 2)) # Adjust figure size as needed

# Hide the axes
ax_area_table.axis('off')
ax_area_table.axis('tight')

# Prepare dataframe for table display
df_area_table_display = df_area[['year', 'Agricultural_Holdings_Area_ha']].copy()
df_area_table_display.columns = ['Year', 'Agricultural Holdings Area (ha)']

# Create the table using matplotlib
table_area = ax_area_table.table(cellText=df_area_table_display.values,
                     colLabels=df_area_table_display.columns,
                     loc='center',
                     cellLoc='center')

# Adjust font size and layout for better readability
table_area.auto_set_font_size(False)
table_area.set_fontsize(10)
table_area.scale(1.2, 1.2) # Scale the table

plt.title('Total Agricultural Holdings Area in Mainland Portugal (1989–2019)', loc='left', fontsize=12, pad=20)

# Save the table as a PNG image
plt.savefig('agricultural_area_table.png', bbox_inches='tight', dpi=300)
print("The table has been saved as 'agricultural_area_table.png'")

plt.close(fig_area_table)
```

#### Figure 3.3 — Evolution of Total Agricultural Area (1989–2019)



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set_style('whitegrid')

fig_area_line = plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='Agricultural_Holdings_Area_ha', data=df_area, marker='o')
plt.title('Evolution of Total Agricultural Area in Mainland Portugal (1989–2019)')
plt.xlabel('Year')
plt.ylabel('Agricultural Holdings Area (ha)')
plt.xticks(df_area['year'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(bottom=0) # Ensure y-axis starts at 0
plt.ticklabel_format(style='plain', axis='y') # Display full numbers on the y-axis
plt.tight_layout()

# Save the plot to a file
plt.savefig('total_agricultural_area_line_chart.png')
print("The line chart has been saved as 'total_agricultural_area_line_chart.png'")

plt.close(fig_area_line) # Close the figure to free up memory
```
